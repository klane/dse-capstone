```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_epochs = 5
default_model_time = sum([1365.1, 1372.6, 1372.4, 1373, 1372.5]) / n_epochs

diffusion_low_time = sum([709.8, 714.9, 714.4, 714.2, 713.3]) / n_epochs
x = [2034, 2042.2]
diffusion_high_time = sum(x) / len(x)

layer_low_time = sum([496.9, 505.8, 505.9, 506.6, 506.4]) / n_epochs
x = [2210.6, 2215.1]
layer_high_time = sum(x) / len(x)

units_low_time = sum([699.5, 710.3, 708, 708.5, 708.3]) / n_epochs
x = [2061.7, 2071.9]
units_high_time = sum(x) / len(x)

diffusion_steps = [1, 2, 3]
diffusion_params = np.array([223745, 372353, 520961]) / 1000
diffusion_time = np.array([diffusion_low_time, default_model_time, diffusion_high_time])
diffusion_loss = []

layer_num = [1, 2, 3]
layer_params = np.array([126209, 372353, 618497]) / 1000
layer_time = np.array([layer_low_time, default_model_time, layer_high_time])
layer_loss = []

units_num = [32, 64, 96]
units_params = np.array([94017, 372353, 835009]) / 1000
units_time = np.array([units_low_time, default_model_time, units_high_time])
units_loss = []
```

```python
plt.figure(figsize=(10, 6))
plt.plot(layer_params, layer_time, units_params, units_time, diffusion_params, diffusion_time)
# plt.title('DCRNN Model Scalability Study', fontsize=12)
plt.legend(['Layers', 'RNN Units', 'Diffusion Steps'], fontsize=12)
plt.xlabel('Model Parameters (x1000)', fontsize=12)
plt.ylabel('Average Epoch Time (s)', fontsize=12)
plt.xlim([0, 900])
plt.ylim([0, 2500])
plt.grid()
plt.savefig('model_scale.png', bbox_inches='tight', dpi=400)
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 3, 1)
plt.plot(layer_num, layer_time, '.-', markersize=10)
plt.xlabel('Number of Layers', fontsize=12)
plt.ylabel('Average Epoch Time (s)', fontsize=12)
plt.ylim([0, 2500])
plt.grid()

plt.subplot(1, 3, 2)
plt.plot(units_num, units_time, '.-', markersize=10)
plt.xlabel('Number of RNN Units', fontsize=12)
plt.ylim([0, 2500])
plt.grid()

plt.subplot(1, 3, 3)
plt.plot(diffusion_steps, diffusion_time, '.-', markersize=10)
plt.xlabel('Diffusion Steps', fontsize=12)
plt.ylim([0, 2500])
plt.grid()

plt.savefig('model_scale_split.png', bbox_inches='tight', dpi=400)
```
